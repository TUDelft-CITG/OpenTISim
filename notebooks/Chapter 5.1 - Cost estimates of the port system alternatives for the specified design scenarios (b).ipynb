```python
# convert data
def PV_func(data):
    if len(data) == 5:
        terminal_capex = data[0]
        terminal_opex = data[1]
        link_capex = 0
        link_opex = 0
        dredging_capex = data[2]
        dredging_opex = data[3]
        ocean_transport = data[4]

    elif len(data) == 10:
        terminal_capex = data[0] + data[1] + data[3]
        terminal_opex = data[2] + data[4]
        link_capex = data[5]
        link_opex = data[6]
        dredging_capex = data[7]
        dredging_opex = data[8]
        ocean_transport = data[9]
        
    elif len(data) == 12:
        terminal_capex = data[0] + data[1] + data[3]
        terminal_opex = data[2] + data[4]
        link_capex = data[5] + data[9]
        link_opex = data[6] + data[10]
        dredging_capex = data[7]
        dredging_opex = data[8]
        ocean_transport = data[11]
  
    else:
        print("pass")
   
    PV_list = [ocean_transport, terminal_capex, terminal_opex, link_capex, link_opex, dredging_capex, dredging_opex]
    bottom_list = [0] + list(np.cumsum(PV_list))
    
    return PV_list, bottom_list
```

```python
# Panamax
PV_panamax_1,  bottom_panamax_1  = PV_func(PV_list_alt1_6000_1000000)

PV_panamax_2a, bottom_panamax_2a = PV_func(PV_list_alt2_20_6000_1000000)
PV_panamax_2b, bottom_panamax_2b = PV_func(PV_list_alt2_40_6000_1000000)
PV_panamax_2c, bottom_panamax_2c = PV_func(PV_list_alt2_60_6000_1000000)

PV_panamax_3a, bottom_panamax_3a = PV_func(PV_list_alt3_20_6000_1000000)
PV_panamax_3b, bottom_panamax_3b = PV_func(PV_list_alt3_40_6000_1000000)
PV_panamax_3c, bottom_panamax_3c = PV_func(PV_list_alt3_60_6000_1000000)

# New-Panamax
PV_new_panamax_1,  bottom_new_panamax_1  = PV_func(PV_list_alt1_12500_1000000)

PV_new_panamax_2a, bottom_new_panamax_2a = PV_func(PV_list_alt2_20_12500_1000000)
PV_new_panamax_2b, bottom_new_panamax_2b = PV_func(PV_list_alt2_40_12500_1000000)
PV_new_panamax_2c, bottom_new_panamax_2c = PV_func(PV_list_alt2_60_12500_1000000)

PV_new_panamax_3a, bottom_new_panamax_3a = PV_func(PV_list_alt3_20_12500_1000000)
PV_new_panamax_3b, bottom_new_panamax_3b = PV_func(PV_list_alt3_40_12500_1000000)
PV_new_panamax_3c, bottom_new_panamax_3c = PV_func(PV_list_alt3_60_12500_1000000)

# ULCS
PV_ULCS_1,  bottom_ULCS_1  = PV_func(PV_list_alt1_21000_1000000)

PV_ULCS_2a, bottom_ULCS_2a = PV_func(PV_list_alt2_20_21000_1000000)
PV_ULCS_2b, bottom_ULCS_2b = PV_func(PV_list_alt2_40_21000_1000000)
PV_ULCS_2c, bottom_ULCS_2c = PV_func(PV_list_alt2_60_21000_1000000)

PV_ULCS_3a, bottom_ULCS_3a = PV_func(PV_list_alt3_20_21000_1000000)
PV_ULCS_3b, bottom_ULCS_3b = PV_func(PV_list_alt3_40_21000_1000000)
PV_ULCS_3c, bottom_ULCS_3c = PV_func(PV_list_alt3_60_21000_1000000)
```

```python
color = cmap_tab20(np.array([15, 0, 1, 2, 3, 4, 5]))
labels = ["Ocean transport", 
          "Terminal capex", "Terminal opex",
          "Transport link capex", "Transport link opex", 
          "Capital dredging", "Maintenance dredging"]

def grouped_stack_costs(values_1a, bottom_1a,
                        values_2a, values_2b, values_2c, bottom_2a, bottom_2b, bottom_2c,
                        values_3a, values_3b, values_3c, bottom_3a, bottom_3b, bottom_3c,
                        width=0.25, alpha=0.6):

    fig, ax = plt.subplots(figsize=(8, 8))
 
    for i in range(len(values_1a)):
        bar = ax.bar([x + 1 * width for x in [0]], values_1a[i], width=width, color=color[i], edgecolor='w', bottom=bottom_1a[i], label=labels[i])
        
    for i in range(len(values_2a)):
        bar = ax.bar([x - 1 * width for x in [1]], values_2a[i], width=width, color=color[i], edgecolor='w', bottom=bottom_2a[i])
        bar = ax.bar([x + 0 * width for x in [1]], values_2b[i], width=width, color=color[i], edgecolor='w', bottom=bottom_2b[i])
        bar = ax.bar([x + 1 * width for x in [1]], values_2c[i], width=width, color=color[i], edgecolor='w', bottom=bottom_2c[i])
        
    for i in range(len(values_2a)):
        bar = ax.bar([x - 1 * width for x in [2]], values_3a[i], width=width, color=color[i], edgecolor='w', bottom=bottom_3a[i])
        bar = ax.bar([x + 0 * width for x in [2]], values_3b[i], width=width, color=color[i], edgecolor='w', bottom=bottom_3b[i])
        bar = ax.bar([x + 1 * width for x in [2]], values_3c[i], width=width, color=color[i], edgecolor='w', bottom=bottom_3c[i])
    
    for alt in [1, 2]:
        label = list(["Onshore", "Barge", "Bridge"])
        ax.text(alt, -0.5*10**9, label[alt], fontsize='large',
            verticalalignment="center", horizontalalignment="center")
           
    plt.xticks([width, 1-width,1,1+width, 2-width, 2, 2+width], ["Onshore", "20 km", "40 km", "60 km", "20 km", "40 km", "60 km"], fontsize='large')

    plt.ylabel("PV of the costs", fontsize='large')
    plt.yticks(np.arange(0, 8*10**9, step = 10**9), ["","$ 1B","$ 2B","$ 3B","$ 4B", "$ 5B", "$ 6B", "$ 7B"])

    plt.legend(title="Categorised cost elements", title_fontsize='large',
        fontsize='large', bbox_to_anchor=(1.0, 0.6), frameon=False)
```

```python
grouped_stack_costs(PV_panamax_1, bottom_panamax_1,
                    PV_panamax_2a, PV_panamax_2b, PV_panamax_2c, bottom_panamax_2a, bottom_panamax_2b, bottom_panamax_2c,
                    PV_panamax_3a, PV_panamax_3b, PV_panamax_3c, bottom_panamax_3a, bottom_panamax_3b, bottom_panamax_3c)
sns.despine()

plt.savefig("SaveFigures/stacked_grouped_costs_post_panamax.png", bbox_inches = 'tight')
```

```python
# Alternative I
onshore_capex        = 4.464402e+08
onshore_opex         = 6.644885e+08
capital_dredging     = 2.054873e+09
maintenance_dredging = 9.260763e+08
ocean_transport      = 7.370249e+08

dredging = [capital_dredging, maintenance_dredging]
terminal = [onshore_capex, onshore_opex]
other = [0, ocean_transport]

total = sum(dredging) + sum(terminal) + sum(other)
percentages = [sum(dredging)/total*100, sum(terminal)/total*100, sum(other)/total*100]
total = 4_850
    
fig, ax = plt.subplots(figsize=(8,8)) 

size = 0.3
vals = np.array([dredging, terminal, other])

cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.array([8, 0, 18]))
inner_colors = cmap(np.array([9, 10, 1, 2, 19, 19]))

outer_wedges, texts = ax.pie(vals.sum(axis=1), radius=1, colors=outer_colors,
                             wedgeprops=dict(width=size, edgecolor='w'))
inner_wedges, texts = ax.pie(vals.flatten(), radius=1-size, colors=inner_colors,
                             wedgeprops=dict(width=size, edgecolor='w'))

outer_labels = ["Dredging", "Terminal", "Ocean transport"]
inner_labels = ["Capital dredging", "Maintenance dredging", "Terminal capex", "Terminal opex","Ocean transport"]

def label(label):
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0)
    kw = dict(arrowprops=dict(arrowstyle="-"),
              bbox=bbox_props, zorder=0, va="center")

    for i, p in enumerate(outer_wedges):
        ang = (p.theta2 - p.theta1)/2. + p.theta1
        y = np.sin(np.deg2rad(ang))
        x = np.cos(np.deg2rad(ang))
        horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
        connectionstyle = "angle,angleA=0,angleB={}".format(ang)
        kw["arrowprops"].update({"connectionstyle": connectionstyle})
        ax.annotate("{}\n{:.1f} %".format(outer_labels[i],percentages[i]), xy=(x, y), xytext=(1.1*np.sign(x), 1.1*y),
                    horizontalalignment=horizontalalignment, fontsize='x-large', **kw)

    ax.legend(inner_wedges, inner_labels,
              title="Legend inner wedges          ",
              title_fontsize='x-large',
              fontsize='x-large',
              loc="center left",
              framealpha=False,
              bbox_to_anchor=(-0.5, 0, 0.5, 1))
    
    ax.set(aspect="equal")
    ax.set_title(label="{}\n\nPV of the costs: $ {:,.0f}M".format(label, total), fontsize='x-large')
    
label('Alternative Onshore')

fig.savefig("SaveFigures/pie_1.png", bbox_inches = 'tight')
```
