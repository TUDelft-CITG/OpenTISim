```python
def plot_trend(maize, soybean, wheat, width, height):
    fig  = plt.figure(figsize=(width, height))
    grid = plt.GridSpec(1, 1, wspace=0.4, hspace=0.5)
    fig1 = fig.add_subplot(grid[0, 0])

    fig1.plot(maize.years, maize.demand    , label='Maize demand')
    fig1.plot(soybean.years, soybean.demand, label='Soybean demand')
    fig1.plot(wheat.years, wheat.demand    , label='Wheat demand')
    fig1.set_title ('Commodity demand')
    fig1.set_xlabel('Year')
    fig1.set_ylabel('Demand [t/y]')
    fig1.legend()

    return fig1
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Linear method

time = []
x1 = []
x2 = []
x3 = []

ini_demand = 1000000
growth1 = 100000
growth2 = 70000
growth3 = 50000

for i in range (20):
    time.append(i)
    x1.append(ini_demand + i*growth1)
    x2.append(ini_demand + i*growth2)
    x3.append(ini_demand + i*growth3)
```

```python
# Exponential method

y1 = []
y2 = []
y3 = []

ini_demand = 1000000
rate1       = 1.01
rate2       = 1.03
rate3       = 1.05

for i in range (20):
    y1.append(ini_demand * (rate1)**i)
    y2.append(ini_demand * (rate2)**i)
    y3.append(ini_demand * (rate3)**i)
```

```python
# Random method

z1 = []
z2 = []
z3 = []

ini_demand = 1000000
base_rate  = 1.02
mu_rate    = 0.02
sigma_rate = 0.015



for i in range (20):
    change = np.random.normal(mu_rate, sigma_rate, 1000)
    rate  =  base_rate + change[0]
    if i == 0:
        demand = ini_demand
    demand = demand * rate
    z1.append(demand)
    
for i in range (20):
    change = np.random.normal(mu_rate, sigma_rate, 1000)
    rate  =  base_rate + change[0]
    if i == 0:
        demand = ini_demand
    demand = demand * rate
    z2.append(demand)
    
for i in range (20):
    change = np.random.normal(mu_rate, sigma_rate, 1000)
    rate  =  base_rate + change[0]
    if i == 0:
        demand = ini_demand
    demand = demand * rate
    z3.append(demand)
```

```python
fig  = plt.figure(figsize=(10, 5))
grid = plt.GridSpec(1, 1, wspace=0.4, hspace=0.5)
fig1 = fig.add_subplot(grid[0, 0])

fig1.plot(time, x1, label='100.000t growth')
fig1.plot(time, x2, label='70.000t growth')
fig1.plot(time, x3, label='50.000t growth')
fig1.set_title ('Commodity demand')
fig1.set_xlabel('Year')
fig1.set_ylabel('Demand [t/y]')
fig1.legend()
```
